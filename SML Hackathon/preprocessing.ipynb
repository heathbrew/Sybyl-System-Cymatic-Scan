# Manual Step of replacing "-" with "s" due to it converting the value to Date format.
# E.g :- 5-22 was being read as 22nd May
# Loading the CSV file which was altered manually

```python
#loading the CSV file that was altered
manual = pd.read_csv(os.getcwd()+"\manual.csv")
```

# Dropping  columns of the Dataset that are not needed

```python
manual = manual.drop(["Unnamed: 0","Unnamed: 0.1"],axis=1)
```

# Updating values of the columns according the model

```python
proven = []
possible = []

for i in manual["Proven victims"]:
    l = list(i.split('s'))
    if(len(l)==1):
        proven.append(int(l[0]))
    else:
        x = int(l[0])
        y = int(l[1])
        proven.append(int((x+y)/2))
        
for i in manual["Possible victims"]:
    l = list(i.split('s'))
    if(len(l)==1):
        possible.append(int(l[0]))
    else:
        x = int(l[0])
        y = int(l[1])
        possible.append(int((x+y)/2))

manual["Proven victims"] = proven
manual["Possible victims"] = possible
```

# Calculating conditional probability to find out crime coefficient 

```python
manual["Poss&Prov"] = abs(manual["Possible victims"] - manual["Proven victims"])
manual["CrimeCofficient"] = round((manual["Poss&Prov"] / manual["Proven victims"]),2)
manual["CrimeCofficient"] = manual["CrimeCofficient"].replace([0.00], 1.00)
manual.drop(["Poss&Prov"],axis=1,inplace=True)
```
